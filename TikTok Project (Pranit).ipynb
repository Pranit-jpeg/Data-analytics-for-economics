You know from the project proposal that the ultimate objective is to use machine learning to classify videos as either claims or opinions. A good first step towards understanding the data might therefore be examining the `claim_status` variable. Begin by determining how many videos there are for each different claim status.


````python
# What are the different values for claim status and how many of each are in the data?
### YOUR CODE HERE ###
# Drop rows where key features are missing (e.g., claim_status or video_transcription_text)
data_cleaned = data.dropna(subset=['claim_status', 'video_transcription_text'])

for col in ['video_view_count', 'video_like_count', 'video_share_count', 'video_download_count', 'video_comment_count']:
    data_cleaned.loc[:, col] = data_cleaned[col].fillna(data_cleaned[col].median())

claim_status_counts = data_cleaned['claim_status'].value_counts()
print("Video counts by claim_status:")
print(claim_status_counts)
````

Next, examine the engagement trends associated with each different claim status.

Start by using Boolean masking to filter the data according to claim status, then calculate the mean and median view counts for each claim status.

````python
# What is the average view count of videos with "claim" status?
### YOUR CODE HERE ###
# Filter for claim videos
claim_data = data_cleaned[data_cleaned['claim_status'] == 'claim']

# Basic statistics for claim videos
print("Claim Videos Analysis")
print("Number of claim videos:", claim_data.shape[0])
print("Mean video view count:", claim_data['video_view_count'].mean())
print("Median video view count:", claim_data['video_view_count'].median())

# Plot a histogram for the view count distribution of claim videos
import matplotlib.pyplot as plt
plt.hist(claim_data['video_view_count'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Video View Count for Claim Videos")
plt.xlabel("Video View Count")
plt.ylabel("Frequency")
plt.show()
````

````python
# What is the average view count of videos with "opinion" status?
### YOUR CODE HERE ###
# Filter for opinion videos
opinion_data = data_cleaned[data_cleaned['claim_status'] == 'opinion']

# Basic statistics for opinion videos
print("Opinion Videos Analysis")
print("Number of opinion videos:", opinion_data.shape[0])
print("Mean video view count:", opinion_data['video_view_count'].mean())
print("Median video view count:", opinion_data['video_view_count'].median())

# Plot a histogram for the view count distribution of opinion videos
plt.hist(opinion_data['video_view_count'], bins=30, color='salmon', edgecolor='black')
plt.title("Distribution of Video View Count for Opinion Videos")
plt.xlabel("Video View Count")
plt.ylabel("Frequency")
plt.show()
````

Use `groupby()` to group the data by `author_ban_status`, then use `agg()` to get the count, mean, and median of each of the following columns:
* `video_view_count`
* `video_like_count`
* `video_share_count`

Remember, the argument for the `agg()` function is a dictionary whose keys are columns. The values for each column are a list of the calculations you want to perform.

````python
### YOUR CODE HERE ###
# Group the data by author_ban_status and aggregate the engagement metrics
engagement_stats = data_cleaned.groupby('author_ban_status').agg({
    'video_view_count': ['count', 'mean', 'median'],
    'video_like_count': ['count', 'mean', 'median'],
    'video_share_count': ['count', 'mean', 'median']
}).reset_index()

# Print the aggregated results
print("Engagement Statistics by Author Ban Status:")
print(engagement_stats)
````

**Question:** What do you notice about the number of views, likes, and shares for banned authors compared to active authors?
* A: Based on previous analyses, we notice that banned authors generally have lower numbers of views, likes, and shares compared to active authors. This could be due to several reasons:
# Reduced Visibility: Videos from banned authors may be suppressed by the platform's algorithms, leading to fewer views.
# Lower Engagement: The reduced exposure may result in fewer likes and shares.
# User Trust: Users might be less inclined to engage with content from banned authors.
By creating the new columns (likes_per_view, comments_per_view, shares_per_view), we can further explore how efficiently each video engages viewers relative to its reach. These metrics provide additional insights into the engagement behavior for both banned and active authors.

Now, create three new columns to help better understand engagement rates:
* `likes_per_view`: represents the number of likes divided by the number of views for each video
* `comments_per_view`: represents the number of comments divided by the number of views for each video
* `shares_per_view`: represents the number of shares divided by the number of views for each video

````python
# Create a likes_per_view column
### YOUR CODE HERE ###
data_cleaned["likes_per_view"] = data_cleaned["video_like_count"] / data_cleaned["video_view_count"]

# Create a comments_per_view column
### YOUR CODE HERE ###
data_cleaned["comments_per_view"] = data_cleaned["video_comment_count"] / data_cleaned["video_view_count"]

# Create a shares_per_view column
### YOUR CODE HERE ###
data_cleaned["shares_per_view"] = data_cleaned["video_share_count"] / data_cleaned["video_view_count"]


print(data_cleaned[["video_view_count", "video_like_count", "video_comment_count", "video_share_count",
                     "likes_per_view", "comments_per_view", "shares_per_view"]].head())

````

Use `groupby()` to compile the information in each of the three newly created columns for each combination of categories of claim status and author ban status, then use `agg()` to calculate the count, the mean, and the median of each group.

````python
### YOUR CODE HERE ###
# Group by claim_status and author_ban_status and aggregate the engagement rate columns
engagement_rate_stats = data_cleaned.groupby(['claim_status', 'author_ban_status']).agg({
    'likes_per_view': ['count', 'mean', 'median'],
    'comments_per_view': ['count', 'mean', 'median'],
    'shares_per_view': ['count', 'mean', 'median']
}).reset_index()

print("Engagement Rate Statistics by claim_status and author_ban_status:")
print(engagement_rate_stats)
````

When building predictive models, the presence of outliers can be problematic. That's the nature of viral content.

Nonetheless, it's good practice to get a sense of just how many of your data points could be considered outliers. The definition of an outlier can change based on the details of your project, and it helps to have domain expertise to decide a threshold. You've learned that a common way to determine outliers in a normal distribution is to calculate the interquartile range (IQR) and set a threshold that is 1.5 * IQR above the 3rd quartile.

In this TikTok dataset, the values for the count variables are not normally distributed. They are heavily skewed to the right. One way of modifying the outlier threshold is by calculating the **median** value for each variable and then adding 1.5 * IQR. This results in a threshold that is, in this case, much lower than it would be if you used the 3rd quartile.

Write a for loop that iterates over the column names of each count variable. For each iteration:

3. Calculate the outlier threshold (median + 1.5 * IQR)

5. Print "Number of outliers, {column name}: {outlier count}"

```

```

### **Task 6. Hypothesis testing**

Before you conduct your hypothesis test, consider the following questions where applicable to complete your code response:

1. Recall the difference between the null hypothesis and the alternative hypotheses. What are your hypotheses for this data project?

*   **Null hypothesis**: There is no difference in number of views between TikTok videos posted by verified accounts and TikTok videos posted by unverified accounts (any observed difference in the sample data is due to chance or sampling variability).

*    **Alternative hypothesis**: There is a difference in number of views between TikTok videos posted by verified accounts and TikTok videos posted by unverified accounts (any observed difference in the sample data is due to an actual difference in the corresponding population means).


Your goal in this step is to conduct a two-sample t-test. Recall the steps for conducting a hypothesis test:


1.   State the null hypothesis and the alternative hypothesis
2.   Choose a signficance level
3.   Find the p-value
4.   Reject or fail to reject the null hypothesis



**$H_0$**: Mean view counts for verified = Mean view counts for unverified.

**$H_A$**: Mean view counts for verified  Mean view counts for unverified.


You choose 5% as the significance level and proceed with a two-sample t-test.

````python
# Conduct a two-sample t-test to compare means
alpha = 0.05

verified_views = data_cleaned[data_cleaned['verified_status'] == 'verified']['video_view_count']
unverified_views = data_cleaned[data_cleaned['verified_status'] == 'not verified']['video_view_count']

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(verified_views, unverified_views, equal_var=False)

print("Two-sample t-test Results:")
print(f"  t-statistic = {t_stat:.4f}")
print(f"  p-value = {p_value:.6f}")

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in average view counts.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant difference in average view counts.")

````
